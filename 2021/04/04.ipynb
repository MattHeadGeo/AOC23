```python
draw = dat_draw
dim = 5
numboards = len(dat_boards)//dim

boards = np.zeros([dim,dim,numboards])
check = np.zeros([dim,dim,numboards])

for i in np.arange(0, numboards):
    boards[:,:,i] = dat_boards[i*dim:(i+1)*dim,:]

count = np.arange(0, numboards)

for n in np.arange(0, len(draw)):
    num = int(draw[n])
    for i in count:
        for j in np.arange(0, dim):
            if np.sum(check[j,:,i]) == 5 or np.sum(check[:,j,i]) == 5:
                boards[check==1] = 0
                if num*np.sum(boards[:,:,i]) > 0:
                    print("board", i+1, "score =", num*np.sum(boards[:,:,i]))
                boards[:,:,i] = 0
            else:
                check[boards==num] = 1
                
```
